### openai

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
QUERY = "You know nothing Jon Snow"
```

```python
from langchain_openai import OpenAIEmbeddings

model = OpenAIEmbeddings(
    model = 'text-embedding-3-large', 
    dimensions = 32)
result = model.embed_query(QUERY)
```

```python
print(str(result))
```

```python
documents = [
    "You know nothing Jon snow",
    "Lannister always pays his debts",
    "Winter is coming"
]

result = model.embed_documents(documents)
```

```python
print(str(result))
```

### hugging face

```python
from langchain_huggingface import HuggingFaceEmbeddings
modelhf = HuggingFaceEmbeddings(
    model_name = 'sentence-transformers/all-MiniLM-L6-v2'
)
resulthf = modelhf.embed_documents(documents)
```

```python
print(str(resulthf))
```
